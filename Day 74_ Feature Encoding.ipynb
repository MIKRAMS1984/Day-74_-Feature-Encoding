# Enkoding time.

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder

le = LabelEncoder()
df['enkoded_time'] = le.fit_transform(df['time'])
print(df.head())
print(df.enkoded_time.value_counts())

```

# Ordinal enkoding of days

```python
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['enkoded_day'] = oe.fit_transform(df[['day']])
df.head()
```

```python
print(df.enkoded_day.value_counts())
print(df.day.value_counts())
```

# Diskretization.

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import KBinsDiscretizer

df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
print(df.age.isnull().sum())
print(df.fare.isnull().sum())
```

```python
# Impute filling values.
df['age'] = df['age'].fillna(df['age'].median())
df['fare'] = df['fare'].fillna(df['fare'].median())
```

```python
print(df.age.isnull().sum())
print(df.fare.isnull().sum())
```

```python
sns.histplot(df.age, kde=True)
```

```python
# age discretization
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
df['age_bin'] = est.fit_transform(df[['age']])
df.head()
```

```python
sns.histplot(df.age_bin, kde=True);
```

```python
sns.histplot(df, x='age', hue='age_bin');
```

# Pandas method for manual binning

```python
df['age_bin2'] = pd.cut(df['age'], bins=3, labels=[0, 1,2])
df.head()
```

```python
sns.histplot(df, x='age', hue='age_bin2', kde=True);
```
